```python
import os
import re
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
```

```python
data = []
with open("../data/txt/all_ravdess_ck.txt", "r") as f:
    for line in f:
       data.append(line.strip())
print(data)
```

```python
ck = {0: 'Happy', 1: 'Angry', 2: 'Disgust', 3: 'Fear', 4: 'Sad', 5: 'Contempt', 6: 'Surprise'}
listvalue = list(ck.values())
listvalue
```

function

```python
def report_pred(datafold, listvalue, eval_fold):
    num = 0
    label_fold = []
    pred_fold = []
    for i in eval_fold:
        data = i.split(", ")
        label = int(data[0][-1])
        pred = int(data[1][-1])
        label_fold.append(label)
        pred_fold.append(pred)
    for i in range(len(datafold)):
        label = int(label_fold[i])
        pred = int(pred_fold[i])
        tf = label == pred
        if not tf:
            print(f'{datafold[i]}, Label: {label} ({listvalue[label]}), Pred: {pred} ({listvalue[pred]}), Correct: {tf}')
            num += 1
    print(f'Number of wrong predictions: {num}')
    return label_fold, pred_fold
        
```

```python
evaldest = "../log/eval_ck4-1/rav/"
normal_data_eval_file = []
glass_data_eval_file = []
for file in os.listdir(evaldest):
    if re.match(r"evalrav4-1", file):
        normal_data_eval_file.append(file)
    else:
        glass_data_eval_file.append(file)
```

```python
eval_normal = []
with open(evaldest+normal_data_eval_file[0], "r") as f:
    for line in f:
        if re.match(r"L", line):
            eval_normal.append(line.strip())
print(eval_normal)
```

```python
eval_glass = []
with open(evaldest+glass_data_eval_file[0], "r") as f:
    for line in f:
        if re.match(r"L", line):
            eval_glass.append(line.strip())
print(eval_glass)
```

```python
label_normal, pred_normal = report_pred(data, listvalue, eval_normal)
```

```python
print(classification_report(label_normal, pred_normal, target_names=listvalue))
```

```python
label_glass, pred_glass = report_pred(data, listvalue, eval_glass)
```
